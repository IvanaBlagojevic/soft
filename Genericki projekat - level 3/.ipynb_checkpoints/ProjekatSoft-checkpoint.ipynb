```python
def prilagodiCifre(cifre):
    raspon = range(0, len(cifre))
    close_kernel = np.ones((5, 5), np.uint8)
    for i in raspon:
        cifra = cifre[i].reshape(28, 28)
        upada = cv2.inRange(cifra, 150, 255)
        closing = cv2.morphologyEx(upada, cv2.MORPH_CLOSE, close_kernel)
        labeled = label(closing)
        regioni = regionprops(labeled)
        if(len(regioni) > 1):
            maksimalna_sirina = 0
            maksimalna_visina = 0
            for j in range(0, len(regioni)):
                pomoc = regioni[j].bbox
                if(pomoc[3] - pomoc[1] > maksimalna_sirina and  pomoc[2] - pomoc[0] > maksimalna_visina):
                    maksimalna_visina = pomoc[3] - pomoc[1]
                    maksimalna_sirina = pomoc[2] - pomoc[0]
                    bbox = pomoc
        else:
            bbox = regioni[0].bbox
        slika_cifre = np.zeros((28, 28))
        x = 0
        for w in range(bbox[0], bbox[2]):
            y = 0
            for h in range(bbox[1], bbox[3]):
                slika_cifre[x, y] = cifra[w, h]
                y += 1
            x += 1
        cifre[i] = slika_cifre.reshape(1, 28*28)
```

```python
#ucitavanje cifara - normalizacija podataka
def ucitaj_podatke():
    with open("train-images.idx3-ubyte",'rb') as f:
        podaci = np.frombuffer(f.read(), np.uint8, offset=16)
        podaci = podaci.reshape(-1,1,28,28)
    return podaci/np.float32(256)

def read_idx(filename):
    with open(filename, 'rb') as f:
        zero, data_type, dims = struct.unpack('>HBB', f.read(4))
        shape = tuple(struct.unpack('>I', f.read(4))[0] for d in range(dims))
        return np.fromstring(f.read(), dtype=np.uint8).reshape(shape)


def ucitaj_labele():
    with open("train-labels.idx1-ubyte", 'rb') as f:
        labele = np.frombuffer(f.read(), np.uint8, offset=8)
    return labele


print("Loading MNIST dataset...")
#pomoc = datasets.fetch_mldata("MNIST original")
#cifre = pomoc.data#.astype('uint8')
#transform(cifre)
#l = pomoc.target
#cifre = x_train
#l = y_train
#nNeighbours = KNeighborsClassifier(n_neighbors=1)
#cifre = ucitaj_podatke()
#l = ucitaj_labele()

cifre = read_idx("train-images.idx3-ubyte")
#cifre = (cifre / 255.0) > 0.5
cifre = np.reshape(cifre, (60000, 28*28))

l = read_idx("train-labels.idx1-ubyte")
prilagodiCifre(cifre)

cifre_test = read_idx("t10k-images.idx3-ubyte")
cifre_test = np.reshape(cifre_test, (10000, 28*28))
l_test = read_idx("t10k-labels.idx1-ubyte")

#HOG
list_hog_fd = []
for cifra in cifre:
    fd = hog(cifra.reshape((28, 28)), orientations=9, pixels_per_cell=(14, 14), cells_per_block=(1, 1), visualise=False)
    list_hog_fd.append(fd)
hog_features = np.array(list_hog_fd, 'float32')

clf = LinearSVC()
clf.fit(hog_features, l)
joblib.dump(clf, "digits_cls.pkl", compress=3)

print('KNN start')
knn = KNeighborsClassifier(n_neighbors=5, algorithm='auto').fit(cifre, l)
print('KNN done')
```

```python
def iznadLinije(x1,y1,x2,y2,sredina):
    import numpy as np
    a = np.array([x1,y1])
    b = np.array([x2,y2])
    p = np.array([sredina[0],sredina[1]])
    # opadajuca prava
    if y1<y2:
        print("Opadajuca prava")
        casa = a
        a = b
        b = casa
    
    d = (sredina[0] - x1)*(y2-y1) - (sredina[1]-y1)*(x2-x1)
    v1 = {x2-x1, y2-y1}
    v2 = {x2-sredina[0], y2-sredina[1]}
    #xp = v1.x*v2.y - v1.y*v2.x  # Cross product
    
    #indikator da se sredina broja nalazi izmedju x-eva linije
    
    iznad = (np.cross(p-a, b-a))
    
    if sredina[0]>a[0] and sredina[0]<b[0]:
        #if sredina[1]<a[1] and sredina[1]>b[1]:
        if d > 0 :
            return True, d  
           # else:
               # return False, d  
        else:
            return False, d
    else:
        return False, d
    
    if iznad < 0 :
        return True, d  
    else:
        return False, d
    
    if a < b:
        pomoc = a
        a = b
        b = pomoc
```

```python
def blizuLinije(x1, y1, x2, y2, sredina):
    duzinaLinije = x2 - x1
    visinaLinije = y1 - y2
    minimalna = 0
    #for i in range(0,duzinaLinije):
     #   for j in range(0,visinaLinije):
      #      razdaljina = sqrt((x1 + i - sredina[0])**2 + (y1-j - sredina[1])**2)
            #print(razdaljina)
       #     if  razdaljina < minimalna:
               # print('Mala razdaljina')
                #print(razdaljina)
         #       razdaljina = minimalna
        #    else:
          #      continue
    import numpy as np
    p2=np.array([x1,y1])
    p1=np.array([x2,y2])
    p3=np.array([sredina[0],sredina[1]])
    
    if y1<y2:
        print("Opadajuca prava blizu linije")
        casa = p1
        p1 = p2
        p2 = casa
    razdaljina=np.cross(p2-p1,p3-p1)/np.linalg.norm(p2-p1)
    #print('Velika razdaljina')
    return razdaljina
```

```python
def tacnoNaPravi(x1,y1,x2,y2,sredina):
    m = (y2-y1)/(x2-x1)
    jednacina = y1 + m*(sredina[0]-x1)
    razlika = abs(sredina[1]-jednacina)
    if razlika < 5:
        return True
    else:
        return False
```

```python
import math
def nadjiElement(element, brojevi, pred):
    pronadjeni = []
    for i,el in enumerate(brojevi):
        eX = element[0]
        eY = element[1]
        x = el[1][0]
        y = el[1][1]
        distanca = math.sqrt(math.pow((x - eX), 2) + math.pow((y - eY), 2))
        if distanca < 17 and el[0]==pred:
            pronadjeni.append(i)
    
    
    return pronadjeni
```

```python
def daLiSamVecSabrala(brojevi, x,y,h,w, novi):
    if len(brojevi) == 0:
        return False
    else:
        for b in (brojevi):
            xStaro = b[1]
            yStaro = b[2]
            hStaro = b[3]
            wStaro = b[4]
            #if (xStaro + wStaro) == (x+w) and (yStaro+hStaro)==(y+h): and  novi==b[0]and    and hStaro==h
            if yStaro < (y + 5) and  xStaro < (x + 5) and w==wStaro :
                 return True
            else:
                continue
    
    return False
```

```python
def display_image(image, color= False):
    if color:
        plt.imshow(image)
    else:
        plt.imshow(image, 'gray')

def invert(image):
    return 255-image

def image_bin(image_gs):
    height, width = image_gs.shape[0:2]
    image_binary = np.ndarray((height, width), dtype=np.uint8)
    ret,image_bin = cv2.threshold(image_gs, 127, 255, cv2.THRESH_BINARY)
    return image_bin

def image_gray(image):
    return cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

def erode(image):
    kernel = np.ones((3,3)) # strukturni element 3x3 blok
    return cv2.erode(image, kernel, iterations=1)

def dilate(image):
    kernel = np.ones((3,3)) # strukturni element 3x3 blok
    return cv2.dilate(image, kernel, iterations=1)
```

```python
def scale_to_range(image): # skalira elemente slike na opseg od 0 do 1
    ''' Elementi matrice image su vrednosti 0 ili 255. 
        Potrebno je skalirati sve elemente matrica na opseg od 0 do 1
    '''
    return image/255
```

```python
def matrix_to_vector(image):
    '''Sliku koja je zapravo matrica 28x28 transformisati u vektor sa 784 elementa'''
    return image.flatten()
```
